# What is Pandas?

Because Python is open source, it has many libraries developed by the community for different purposes. Pandas has emerged as the standard library for data manipulation and analysis. Often times, a search provides many suggestions for doing a certain data task in Python. While base Python has data capabilities, Pandas is the preferred library for data work.

# Reading Data

This section introduces some easy ways to read data into Python Pandas. While there are many options, I find these to be the simplest.

The data files that Pandas can directly import should be in .csv format, Excel .xlsx, or Stata .dta format. There is no need to save data in a Python-specific format.

## Uploading files

When using Google Colab notebooks, you first need to upload your data files into the runtime.

Click on the "Folders" icon on the left sidebar and drag-and-drop your CSV, Excel, or Stata file into the window. Alternatively, you can click on the upload bottun and select the file in a dialog box. This will upload the file into the Colab runtime.

Screenshot from 2023-11-08 07-48-07.png

Also, please note that read_stata cannot read zip files. If your Stata file is compressed, you should extract it first and then upload the dta file.

## read_stata

Once the file is uploaded and the libraries you are going to use imported, you can read it into a DataFrame.

```python
data = pd.read_stata('file.dta')
```

Make sure the filename matches the uploaded file.

You can also reference the full Google Colab filesystem path if needed:

```
data = pd.read_stata('/content/file.dta')
```

For example, we can import a dta file of survey data downloaded from a website:

To read a dta file into a DataFrame:

```python
data = pd.read_stata('LFS-71M0001-E-2023-August_F1.dta')
```

By default Pandas will use the dta filename as the DataFrame name. You can also specify a custom name.

When you are reading the dta files, you can indicate if you need the values(convert_categoricals = False) or the lables(convert_categoricals = True) for the categorical variables as below. The read_stata by default read the lables not the values.

```python
data = pd.read_stata('LFS-71M0001-E-2023-August_F1.dta', convert_categoricals=False)
```

Then to read Excel:

```python
data = pd.read_excel('data.xlsx')
```

Or CSV:

```python
data = pd.read_csv('data.csv')
```

Pandas will automatically convert the data into a DataFrame.

One difference from R is that Pandas imports will directly read into DataFrame format, whereas R distinguishes between tibble and data frame formats.

## Basic Regression Analysis


We have to import another library for regression.

```python
import statsmodels.formula.api as smf
```

````python
import statsmodels.formula.api as smf
````

##Univariate regression

````python
# Drop missing values from the columns used in the regression
mydata_clean = data.dropna(subset=['HRLYEARN', 'TENURE'])

# Create the model
model = smf.ols('HRLYEARN ~ TENURE', data=mydata_clean)

# Fit the model
results = model.fit()

# Display regression results
print(results.summary())
````

## Log transformation

````python
import numpy as np

# Create the model
model = smf.ols('np.log(HRLYEARN) ~ TENURE', data=mydata_clean)

# Fit the model
results = model.fit()

# Display regression results
print(results.summary())
````

## Multivariate regression

````python
# Create the model
model = smf.ols('HRLYEARN ~ TENURE + ESTSIZE', data=mydata_clean)

# Fit the model
results = model.fit()

# Display regression results
print(results.summary())
````

# Appendix

Here are some commonly used Pandas commands:

- `read_csv()` - read CSV file into DataFrame
- `read_excel()` - read Excel file into DataFrame
- `read_sql()` - read SQL database table into DataFrame

**View data**

- `head()` - view first N rows
- `tail()` - view last N rows
- `sample()` - sample random rows

**Select data**

- `[]` - select columns
- `.loc[]` - select rows by label
- `.iloc[]` - select rows by position

**Summarize data**

- `describe()` - summary statistics
- `groupby()` - group by one or more columns
- `agg()` - aggregate after grouping
- `pivot_table()` - pivot table from data

**Merge data**

- `merge()` - merge DataFrames on column(s)
- `join()` - join DataFrames on index
- `concat()` - concatenate DataFrames  

**Statistics**

- `corr()` - correlation
- `cov()` - covariance

**Modeling**

- `ols()` - ordinary least squares regression
